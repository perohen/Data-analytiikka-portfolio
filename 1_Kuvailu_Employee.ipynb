<b>Uusien muuttujien laskeminen</b>
<ul>
    <li>Lasken uuden <i>AgeGroup</i>-muuttujan, jossa työntekijöiden iät luokitellaan ikäluokkiin.</li>
    <li>Lasken uuden <i>MonthlyIncomeGroup</i>-muuttujan, jossa työntekijöiden kuukausipalkat luokitellaan palkkaluokkiin.</li>
</ul>

```python
bins = [18, 28, 38, 48, 58, 68]
df['AgeGroup'] = pd.cut(df['Age'], bins = bins)
df.head()
```

```python
bins = [0,1000,3000,5000,7000,9000,11000,13000,15000,17000,19000,21000]
df['MonthlyIncomeGroup'] = pd.cut(df['MonthlyIncome'], bins = bins)
df.head()
```

<b>Crosstab()</b>-funktiolla voin havainnollistaa ikäluokkien kokoa lukumäärän ja prosenttimäärän avulla. Henkilöstöstä enemmistö on 28–37-vuotiaita.

```python
df3 = pd.crosstab(df['AgeGroup'], 'n')
df3.columns.name = ''
df3['%'] = df3/df3.sum()*100

df3.style.format({'%': '{:.1f} %'})
```

```python
# width=1 laittaa pylväät kiinni toisiinsa
# rot=45 kääntää x-akselin ticksien nimiöitä 45 astetta
ax1 = df3['n'].plot.bar(width=1, rot=45, legend=False, edgecolor='black', color='C0')

ax1.set_ylabel('count')
```

Lisäksi voin havainnollistaa kuukausituloja vastaavalla tavalla. Enemmistö henkilöstöstä saa kuukausipalkkaa 1 000–2 999 euroa.

```python
df4 = pd.crosstab(df['MonthlyIncomeGroup'], 'n')
df4.columns.name = ''
df4['%'] = df4/df4.sum()*100

df4.style.format({'%': '{:.1f} %'})
```

```python
# width=1 laittaa pylväät kiinni toisiinsa
# rot=45 kääntää x-akselin ticksien nimiöitä 45 astetta
ax2 = df4['n'].plot.bar(width=1, rot=45, legend=False, edgecolor='black', color='C0')

ax2.set_ylabel('count')
```

<p>Voin havainnollistaa työntekijöiden koulutustaustaa taulukkomuodossa ja kuvioina. Lisäksi voin ottaa sukupuolten väliset erot koulutuksessa tarkasteluun.</p>

```python
education = ['Below College', 'College', 'Bachelor', 'Master', 'Doctor']
df5 = pd.crosstab(df['Education'], 'n')
df5.index = education
df5.columns.name = ''
df5['%'] = df5/df5.sum()*100
df5.style.format({'%': '{:.1f} %'})
```

```python
ax3 = df5['%'].plot.barh()
n = df5['n'].sum()
ax3.set_xlabel('n=' + str(n))
ax3.xaxis.set_major_formatter(myFmt100)
```

```python
df6 = pd.crosstab(df['Education'], df['Gender'])
df6.index = education
df6
```

```python
# Ristiintaulukointi, prosentit (normalize) sarakkeiden mukaan
df7 = pd.crosstab(df['Education'], df['Gender'], normalize = 'columns')
df7.index = education # Riviotsikot koulutus-listasta
# N-arvot mukaan
miehet = df['Education'][df['Gender']=='Male'].count()
naiset = df['Education'][df['Gender']=='Female'].count()
df7.columns = ['Female, n=' + str(naiset),'Male, n=' + str(miehet)]
(df7*100).style.format('{:.1f} %')
```

```python
# Edellinen pylväinä
ax4 = df7.plot.barh(legend='reverse')
ax4.set_xlabel('Percent of Gender')
ax4.xaxis.set_major_formatter(myFmt1)
```

```python
# Edellinen 100 % pinottuina vaakapylväinä
ax5 = df7.T.plot.barh(stacked = True)
ax5.set_xlabel('Percent of Gender')
ax5.legend(loc=(-0.15, -0.25), ncol=5)
ax5.xaxis.set_major_formatter(myFmt1)
```

<p>Voin koota useiden muuttujien frekvenssit yhteen taulukkoon ja havainnollistaa niitä pylväskuvioina. Pythonissa tämä sujuu helposti <b>value_counts()</b>-funktion avustuksella.</p>

```python
df8 = df['EnvironmentSatisfaction'].value_counts(sort = False, normalize = True).to_frame()
df8['JobInvolvement'] = df['JobInvolvement'].value_counts(sort = False, normalize = True)
df8['JobSatisfaction'] = df['JobSatisfaction'].value_counts(sort = False, normalize = True)
df8['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].value_counts(sort = False, normalize = True)
satisfaction = ['Low', 'Medium', 'High', 'Very High']
df8.index = satisfaction
df8.loc['Total'] = df8.sum()
(df8*100).style.format('{:.1f} %')
```

```python
# Edellinen 100 % pylväinä
ax6 = df8.drop('Total').T.plot.barh(stacked=True, color=['#C44E52','#D65F5F','#4878CF','#4C72B0'])
ax6.legend(loc=(0.04, 1.1), ncol=4)
ax6.xaxis.set_major_formatter(myFmt1)
```

```python
axes = (df8.drop('Total')*100).plot.bar(subplots=True, layout=(1, 4), sharex=True, sharey=True, 
              figsize=(10, 2), color='maroon', legend=False)
fig = plt.gcf()
fig.text(0.08, 0.5, "%")
```
